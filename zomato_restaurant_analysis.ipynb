```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
df = pd.read_csv("zomato.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df['rate'].unique()
```

```python
df['rate'] = df['rate'].replace(
    ['NEW','-'],
    np.nan
)
```

```python
df['rate'] = df['rate'].str.split('/').str[0]
```

```python
df['rate'] = pd.to_numeric(df['rate'])
```

```python
df['rate'].head()
```

```python
df['approx_cost(for two people)'] = (
    df['approx_cost(for two people)']
    .astype(str)
    .str.replace(',','')
)
```

```python
df['approx_cost(for two people)'] = pd.to_numeric(
    df['approx_cost(for two people)'],
    errors='coerce'
)
```

```python
plt.figure(figsize=(8,5))

df['rate'].hist(
    bins=20
)

plt.title('Restaurant Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')

plt.show()
```

```python
top_locations = (
    df['location']
    .value_counts()
    .head(10)
)

top_locations
```

```python
df['online_order'].value_counts()
```

```python
df['online_order'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Online Order Availability')
plt.ylabel('')

plt.show()
```

```python
df['book_table'].value_counts()
```

```python
df['book_table'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Table Booking Availability')
plt.ylabel('')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='votes',
    y='rate'
)

plt.title('Votes vs Ratings')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['approx_cost(for two people)'],
    bins=30
)

plt.title('Cost Distribution')

plt.show()
```

```python
df.to_csv(
    'zomato_cleaned.csv',
    index=False
)
```
